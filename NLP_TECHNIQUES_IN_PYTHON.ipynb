```python
#tokenizaton
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
text = "Hello, I am ABC. This is a python lab of 2 HOURS."
sentences = sent_tokenize(text)
print(sentences)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
word_tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_sentence = [w for w in word_tokens if w.lower() not in stop_words and w.isalpha()]

print("Original text:", text)
print("Filtered sentence (stopwords removed):")
print(filtered_sentence)
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed_words = [ps.stem(w) for w in filtered_sentence]

print("Filtered sentence:", filtered_sentence)
print("Stemmed words:", stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_sentence]

print("Filtered sentence:", filtered_sentence)
print("Stemmed words:", stemmed_words)
print("Lemmatized words:", lemmatized_words)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')

pos_tags = nltk.pos_tag(word_tokens)
print("Original text:", text)
print("POS Tags:", pos_tags)
```

```python
from collections import Counter
from nltk.tokenize import word_tokenize
current_word_tokens = word_tokenize(text)
word_frequencies = Counter(current_word_tokens)
print("\nWord Frequencies:")
for word, freq in word_frequencies.most_common(10):
    print(f"'{word}': {freq}")
```
